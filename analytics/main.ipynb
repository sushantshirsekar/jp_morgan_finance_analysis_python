## Task 1 : 
#### 1. Remove/treat any special characters or non-numeric entries from financial fields.

```python
# Checking data type of all numeric fields.
print(main_df['AccountBalance'].dtype)
print(main_df['TransactionAmount'].dtype)
print(main_df['RiskScore'].dtype)
print(main_df['CreditRating'].dtype)
print(main_df['TenureMonths'].dtype)
```

```python
# Treated special characters or non-numeric entries from financial fields.
# Converted currency amounts into numerical format.

main_df['TransactionAmount'] = main_df['TransactionAmount'].replace('[₹$,]', '', regex=True).astype(float)
main_df['AccountBalance'] = main_df['AccountBalance'].replace('[₹$,]', '', regex=True).astype(float)
main_df['RiskScore'] = main_df['RiskScore'].astype(float)
```

```python
# Converted whole number numeric format columns (CreditRating, TenureMonths) to integer. 

main_df['CreditRating'] = main_df['CreditRating'].astype('int64')
main_df['TenureMonths'] = main_df['TenureMonths'].astype('int64')
```

```python
main_df.info()
```

```python
# Handling Missing Values

main_df = main_df.dropna(subset=['AccountBalance'])
```

```python
main_df.info()
```

```python
# Validatted and Formatted (Default Format) Date Columns.
main_df['TransactionDate'] = pd.to_datetime(main_df['TransactionDate'], errors='coerce')
```

```python
main_df.info()
```

```python
# Converted all string format columns to a single string format type (capitalized).

main_df['AccountType'] = main_df['AccountType'].str.strip().str.capitalize()
main_df['TransactionType'] = main_df['TransactionType'].str.strip().str.capitalize()
main_df['Product'] = main_df['Product'].str.strip().str.capitalize()
main_df['Firm'] = main_df['Firm'].str.strip().str.capitalize()
main_df['Region'] = main_df['Region'].str.strip().str.capitalize()
main_df['Manager'] = main_df['Manager'].str.strip().str.capitalize()
```
